```python
#Vector Accumulator!
from pyspark.accumulators import AccumulatorParam
class VectorAccumulatorParam(AccumulatorParam):
    def zero(self, value):
        return [0.0] * len(value)
    def addInPlace(self, val1, val2):
        for i in range(len(val1)):
             val1[i] += val2[i]
        return val1

va = sc.accumulator(np.zeros(40), VectorAccumulatorParam())
print(va.value)
```

```python
#Helper Functions
def convertToInt(rowIn):
    output = []
    for j, item in enumerate(rowIn):
        if item == '':
            output.append(np.nan)
        else:
            if j < 14:
                output.append(int(item)) 
            else:
                output.append(item)
    return output

def countNulls(rowIn, accum):
    for j, item in enumerate(rowIn):
        if item == 'null':
            a = np.zeros(40)
            a[j] += 1
            accum += a
```

```python
#Parse Data
if 2 in runOptions:
    parsedTrain = trainRDD.map(lambda x: (x.split('\t'))) \
        .map(convertToInt)
    
    print(parsedTrain.take(2))
```

```python
#Count Nulls
if 3 in runOptions:
    #Reset accumulator
    va = sc.accumulator(np.zeros(40), VectorAccumulatorParam())

    #Run our null counting function
    parsedTrain.foreach(lambda x: countNulls(x, va))
    numNulls = va.value
    
    print('Number of Null Values by Column:\n', numNulls)
    print('\nPercent of Null Values by Column:\n', (numNulls / trainRowCount))
    for item in numNulls:
        print('{:.2%}'.format(item/trainRowCount), end=' | ')
```

```python
#Check label balance / imbalance
if 3 in runOptions:
    posLabelCount = parsedTrain.map(lambda x: x[0]).sum()
    print('Positive Count:', posLabelCount)
    print('Negative Count:', (trainRowCount - posLabelCount))
    print('Percent Positive:', (posLabelCount / trainRowCount))
```

```python
def analyzeUnique(var, colNum):
    """
    Function to count unique values in a column, 
    and then plot the distribution up to the top 100 by count
    """
    print('CATEGORICAL VARIABLE {}:'.format(var))
    uniqueCount = parsedTrain.map(lambda x: ( (x[0], x[colNum]), 1)) \
        .reduceByKey(lambda x,y: x+y) \
        .sortBy(lambda x: -x[1])

    #Number of unique values
    numCategories = uniqueCount.map(lambda x: (x[0][0], x[0][1], x[1])) \
        .map(lambda x: x[1]) \
        .distinct().count()

    print('Unique Values:', numCategories)
    print('Top 10 by Count:', uniqueCount.take(10))

    #Plot Histogram, by label
    print('Distribution:')
    df = pd.DataFrame(uniqueCount.map(lambda x: (x[0][0], x[0][1], x[1])).take(100))
    df.columns = ['Label', 'Category', 'Count']
    df.pivot('Category','Label') \
        .sort_values(('Count',0), ascending=False) \
        .plot(kind='bar', stacked=True)
        #Note, without limiting it to plot the top X, it will fail with large X.
    
    plt.title('Histogram of Variable {} by Label'.format(var))
    plt.show()
    print() #surpresses output
```

```python
#Get unique values ... need to get for Label, and C1-26
if 4 in runOptions:
    for colNum, var in enumerate(colLabels):
        if var[0] == 'C':  #Only run for categorical variables
            analyzeUnique(var, colNum)
```
